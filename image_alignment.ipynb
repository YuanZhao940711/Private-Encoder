```
import glob

from PIL import Image
from tqdm import tqdm
```

```
IMG_EXTENSIONS = [
    '.jpg', '.JPG', '.jpeg', '.JPEG',
    '.png', '.PNG', '.ppm', '.PPM', '.bmp', '.BMP', '.tiff'
]

def is_image_file(filename):
  return any(filename.endswith(extension) for extension in IMG_EXTENSIONS)
  
```

```
input_path = "./image_datasets/ffhq10k_11k"
files_path = []
for root, dirs, files in os.walk(input_path):
  for file_name in files:
    if is_image_file(file_name):
      file_path = os.path.join(root, file_name)
      files_path.append(file_path)
files_path = sorted(files_path)

output_path = input_path + "_aligned"
os.makedirs(output_path, exist_ok=True)
```

```
import dlib
from scripts.align_all_parallel import align_face

predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

counts = 0
try:
  with tqdm(files_path, total=len(files_path)) as t:
    for img_path in t:
      try:
        aligned_image = align_face(filepath=img_path, predictor=predictor)
        aligned_image.save(os.path.join(output_path, os.path.basename(img_path)))
      except:
        continue
      counts += 1
except KeyboardInterrupt:
  t.close()
  raise

print("\n{} images have been aligned!".format(counts))
```
