```python
def resize_img(img_path, resolution, save_path):
  img_in = Image.open(img_path)
  img_out = img_in.resize(resolution, Image.ANTIALIAS)
  img_out.save(save_path)
  print("Resize input image {} and export to {}".format(os.path.basename(img_path),save_path))

resize_img(img_path='image_datasets/Comparison_Images/10046_11.png',resolution=(256,256), save_path='image_datasets/Comparison_Images/10046_11_resized.png')
resize_img(img_path='image_datasets/Comparison_Images/10840_11.png',resolution=(256,256), save_path='image_datasets/Comparison_Images/10840_11_resized.png')
```

```python
IMG_EXTENSIONS = [
    '.jpg', '.JPG', '.jpeg', '.JPEG',
    '.png', '.PNG', '.ppm', '.PPM', '.bmp', '.BMP', '.tiff'
]
def is_image_file(filename):
  return any(filename.endswith(extension) for extension in IMG_EXTENSIONS)


def resize_by_width(infile, image_size):
    """按照宽度进行所需比例缩放"""
    im = Image.open(infile)
    (x, y) = im.size
    lv = round(x / image_size, 2) + 0.01
    x_s = int(x // lv)
    y_s = int(y // lv)
    print("New Resolution is", x_s, y_s)
    out = im.resize((x_s, y_s), Image.ANTIALIAS)
    return out
def get_new_img_xy(infile, image_size):
    """返回一个图片的宽、高像素"""
    im = Image.open(infile)
    (x, y) = im.size
    lv = round(x / image_size, 2) + 0.01
    x_s = x // lv
    y_s = y // lv
    # print("x_s", x_s, y_s)
    # out = im.resize((x_s, y_s), Image.ANTIALIAS)
    return x_s, y_s
# 定义图像拼接函数
def image_compose(image_size, col_num, row_num, image_names, image_save_path, x_new, y_new):
    to_image = Image.new('RGB', (col_num * x_new, row_num * y_new))  # 创建一个新图
    # 循环遍历，把每张图片按顺序粘贴到对应位置上
    total_num = 0
    for y in range(1, row_num + 1):
        for x in range(1, col_num + 1):
            from_image = resize_by_width(image_names[col_num * (y - 1) + x - 1], image_size)
            # from_image = Image.open(image_names[image_colnum * (y - 1) + x - 1]).resize((image_size,image_size ), Image.ANTIALIAS)
            to_image.paste(from_image, ((x - 1) * x_new, (y - 1) * y_new))
            total_num += 1
            if total_num == len(image_names):
                break
    print(image_save_path)
    to_image.save(image_save_path)

def merge_images(root_path, image_size, col_num):
    assert os.path.isdir(root_path), '{} is not a valid directory'.format(images_path)
    files_name = sorted(os.listdir(root_path))
    images_path = []
    for file_name in files_name:
      if is_image_file(file_name):
        file_path = os.path.join(root_path, file_name)
        images_path.append(file_path)
    images_path = sorted(images_path)
    print("{} input images\nCorresponded path are: {}".format(len(images_path), images_path))

    save_path = '{}.png'.format(root_path)

    row_res = len(images_path) % col_num
    if row_res == 0:
        row_num = len(images_path) // col_num
    else:
        row_num = len(images_path) // col_num + 1

    x_list = []
    y_list = []
    for img_file in images_path:
        img_x, img_y = get_new_img_xy(img_file, image_size)
        x_list.append(img_x)
        y_list.append(img_y)
    #print("x_list", sorted(x_list))
    #print("y_list", sorted(y_list))

    x_new = int(x_list[len(x_list) // 5 * 4])
    y_new = int(x_list[len(y_list) // 5 * 4])

    image_compose(image_size, col_num, row_num, images_path, save_path, x_new, y_new)  # 调用函数
```

```python
image_dir_path = 'image_datasets/Comparison_Images'  # 图片集地址
image_size = 256    # 每张小图片的大小
image_colnum = 12   # 合并成一张图后，一行有几个小图
merge_images(image_dir_path, image_size, image_colnum)
```
