```
from google.colab import drive
drive.mount('/content/drive')

import os
os.chdir('/content/drive/My Drive/pSp_self_customize')

from tqdm import tqdm
from PIL import Image
```

```
IMG_EXTENSIONS = [
    '.jpg', '.JPG', '.jpeg', '.JPEG',
    '.png', '.PNG', '.ppm', '.PPM', '.bmp', '.BMP', '.tiff'
]

def is_image_file(filename):
  return any(filename.endswith(extension) for extension in IMG_EXTENSIONS)

def image_addressing(input_path):
  files_path = os.listdir(input_path)
  images_fullpath = []
  for file_name in files_path:
    file_path = os.path.join(input_path, file_name)
    if is_image_file(file_path):
      images_fullpath.append(file_path)
    else:
      images_partialpath = image_addressing(file_path)
      images_fullpath.extend(images_partialpath)
  return images_fullpath

def path_transfer(source_path, output_path):
  os.makedirs(output_path, exist_ok=True)
  images_path = image_addressing(source_path)
  print("\nFind {} images!\n".format(len(images_path)))
  
  try:
    with tqdm(images_path, total=len(images_path)) as t:
      for path in t:
        img = Image.open(path)
        img.save(os.path.join(output_path, os.path.basename(path)))
        #os.remove(path) # delete original image
  except KeyboardInterrupt:
    t.close()
    raise
```

```
source_path = './image_datasets/original_folders'
output_path = './image_datasets/original_images'
path_transfer(source_path=source_path, output_path=output_path)
```
